```python
from deepchecks.nlp import TextData
from deepchecks.nlp import checks
from deepchecks.nlp.checks import TextEmbeddingsDrift
import pandas as pd
```

```python
perturbation = "gaussiannoise"
strength = 10
ref_path = "/home/vishesh/Desktop/datasets/fleurs/embeddings/train.csv"
# cur_path = f"/home/vishesh/Desktop/datasets/fleurs/embeddings/perturbations/{perturbation}/test_results_{perturbation}_{strength}.csv"
cur_path = f"/home/vishesh/Desktop/datasets/fleurs/embeddings/test.csv"
```

```python
reference = pd.read_csv(ref_path)
current = pd.read_csv(cur_path)
# current.drop_duplicates(subset=['Predictions'], inplace=True)
# reference.drop_duplicates(subset=['Predictions'], inplace=True)
```

```python
train_pred = reference["Predictions"]
test_pred = current["Predictions"]
train_embeddings = reference.iloc[:, -768:].to_numpy()
test_embeddings = current.iloc[:, -768:].to_numpy()
```

```python
# Initialize the TextData object
train_text_data = TextData(train_pred)
test_text_data = TextData(test_pred)

# Set the embeddings:
train_text_data.set_embeddings(train_embeddings)
test_text_data.set_embeddings(test_embeddings)

# Run the check
TextEmbeddingsDrift().run(train_text_data, test_text_data)
```
